# 05 · Optimización de Pricing

Objetivo: evaluar escenarios de precio ($120, $180, $240) con break-even, margen y sensibilidad.

```python
from pathlib import Path
import sys

# Resolver raíz del proyecto de forma robusta

def find_root(start=None):
    p = Path(start or Path.cwd()).resolve()
    for _ in range(6):
        if (p / 'requirements.txt').exists() and (p / 'src').exists():
            return p
        if (p / '.git').exists() and (p / 'src').exists():
            return p
        p = p.parent
    return Path.cwd()

ROOT = find_root()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.pricing_optimization import simulate_pricing_scenarios
from src.visualization import plot_pricing_results
import pandas as pd

prices = [120, 180, 240]
# Supuestos iniciales (ajustables):
adoption = {120: 485, 180: 350, 240: 180}
fixed_costs = 25000  # marketing + sistema + overhead año 1
variable_cost_per_customer = 80  # soporte, beneficios, costos variables

results = simulate_pricing_scenarios(prices, adoption, fixed_costs, variable_cost_per_customer)
print(pd.DataFrame(results))
fig = plot_pricing_results(results)
fig.show()
```
